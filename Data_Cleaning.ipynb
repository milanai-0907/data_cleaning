```python
import re

def extract_age(age):
    age_num = re.findall('[0-9]+', str(age))

    if len(age_num) > 0:
        return age_num[0]
    else:
        return age

df['Age'] = df['Age'].apply(lambda x: extract_age(x))
```

```python
df
```

```python
df['Age'].unique()
```

```python
#we storted all the age values except the nan years in the variable df_age
df_age=df[df['Age']!='nan years']['Age']
```

```python
age_median=int(df_age.dropna().astype('int64').median())
```

```python
print(age_median)
```

```python
#we replace nan values with median
df.replace('nan years',age_median,inplace=True)
```

```python
df['Age'].unique()
```

```python
#replace null values withmedian
df['Age'].fillna(age_median,inplace=True)
#df['Age']=df['Age'].fillna(age_median)
```

```python
df.isnull().sum()
```

```python
#we fill the null value of purchse amount
df['Purchase_Amount'].fillna(df['Purchase_Amount'].median(),inplace=True)
```

**Handling Outliers**

```python
import seaborn as sns
import matplotlib.pyplot as plt
cols=['Age','Purchase_Amount','Feedback_Score']
plt.figure(figsize=(12,4))
for i,col in enumerate(cols,1):
  plt.subplot(1,3,i)
  sns.boxplot(y=df[col],color='red',width=0.3)
  plt.title(f'Boxplot for {col}',fontsize=12)
  plt.ylabel('')
  plt.grid(axis='y',linestyle='--',alpha=0.6)
plt.tight_layout
plt.show()
```

```python
from scipy import stats
import numpy as np
z_score=np.abs(stats.zscore(df[['Age','Purchase_Amount']]))
print(z_score)
```

```python
df[(z_score>3).any(axis=1)]
```

```python
df_clean=df[~(z_score>3).any(axis=1)]
```
