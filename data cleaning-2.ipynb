```python
#handiling missing values
#we will drop the rows which missing the Employee_ID
df.dropna(subset=['Employee_ID'],inplace=True)
df
```

```python
df['Years_of_Experience'].unique()
```

```python
import re

def extract_experience(Years_of_Experience):
    experience_num = re.findall('[0-9]+', str(Years_of_Experience))

    if len(experience_num) > 0:
        return experience_num[0]
    else:
        return Years_of_Experience

df['Years_of_Experience'] = df['Years_of_Experience'].apply(lambda x: extract_experience(x))
```

```python
df
```

```python
df['Years_of_Experience'].unique()
```

```python
#we storted all the Years_of_Experience values except the nan years in the variable df_Years_of_Experience
df_Years_of_Experience=df[df['Years_of_Experience']!='nan years']['Years_of_Experience']
```

```python
Years_of_Experience_median=int(df_Years_of_Experience.dropna().astype('float').median())
```

```python
print(Years_of_Experience_median)
```

```python
#we replace nan values with median
df.replace('nan years',Years_of_Experience_median,inplace=True)
```

```python
df['Years_of_Experience'].unique()
```

```python
#replace null values withmedian
df['Years_of_Experience'].fillna(Years_of_Experience_median,inplace=True)
```

```python
df.isnull().sum()
```

```python
#we fill the null value of Monthly_Salary
df['Monthly_Salary'].fillna(df['Monthly_Salary'].median(),inplace=True)
```

```python
df['Performance_Rating']
```

```python
#we will use mode here to fill the null value to fix the null value as using the median would get
df['Performance_Rating'].fillna(df['Performance_Rating'].mode()[0],inplace=True)
```

```python
cols=['Years_of_Experience','Monthly_Salary','Performance_Rating']
plt.figure(figsize=(12,4))
for i,col in enumerate(cols,1):
  plt.subplot(1,3,i)
  sns.boxplot(y=df[col],color='red',width=0.3)
  plt.title(f'Boxplot for {col}',fontsize=12)
  plt.ylabel('')
  plt.grid(axis='y',linestyle='--',alpha=0.6)
plt.tight_layout
plt.show()
```

```python
from scipy import stats
import numpy as np
z_score=np.abs(stats.zscore(df[['Years_of_Experience','Monthly_Salary','Performance_Rating']]))
print(z_score)
```
